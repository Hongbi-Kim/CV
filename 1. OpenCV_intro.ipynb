### 영상 읽기

```python
img = cv2.imread('./fig/puppy.bmp',
                 flags = cv2.IMREAD_COLOR)  # default

# retval: numpy.ndarray로 반환
print(type(img))
print('image dimension = ',img.shape)
```

```python
img2 = cv2.imread('./fig/puppy.bmp',
                 flags = cv2.IMREAD_GRAYSCALE)
print(type(img))
img2.shape 
```

- flags:
    - cv2.IMREAD_COLOR: BGR color로 읽기
    - cv2.IMREAD_GRAYSCALE: 그레이 color로 읽기
    - cv2.IMREAD_UNCHANGED: 파일 속성대로 읽기
    - cv2.IMREAD_REDUCED_GRAYSCALE_2 
    - cv2.IMREAD_REDUCED_COLOR_2 

```python
img = cv2.imread('./fig/puppy.bmp', flags=cv2.IMREAD_GRAYSCALE)

if img is None:
    print('image read failed')
    sys.exit()

cv2.namedWindow('image')
cv2.imshow('image', img)
cv2.waitKey()

cv2.destroyAllWindows()
# BGR 순서
```

### image 저장하기: cv2.imwrite()

```python
img = cv2.imread('./fig/puppy.bmp', flags=cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()
print(img.shape)

# resize(src, dsize[,dst[,fx[, fy[, interpolation]]]]) -> dst
# interpolation : 영상 보관 방법
img_re = cv2.resize(img, (320, 240), interpolation = cv2.INTER_AREA)

cv2.namedWindow('image')
cv2.namedWindow('image_re')
cv2.imshow('image', img)
cv2.imshow('image_re', img_re)

cv2.imwrite('./fig/puppy_resize.png', img_re)

cv2.waitKey()
cv2.destroyAllWindows()
```

### matplotlib 라이브러리를 이용한 영상출력

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image')
cv2.imshow('image', img) # bgr로 인식

plt.imshow(img) # rgb로 인식
plt.axis('off')
plt.show()

while True:
    if cv2.waitKey() == 27:
        break

cv2.destroyAllWindows()
```

```python
imgRGB = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image')
cv2.imshow('image', img) # bgr로 인식

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(imgRGB) # rgb로 인식
plt.axis('off')
plt.show()

while True:
    if cv2.waitKey() == 27:
        break

cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image')
cv2.imshow('image', img) # bgr로 인식

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize = (12,6))
plt.subplot(131), plt.imshow(img), plt.axis('off')
plt.subplot(132), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(133), plt.imshow(imgGray, cmap='gray'), plt.axis('off')
plt.show()

while True:
    if cv2.waitKey() == 27:
        break

cv2.destroyAllWindows()
```

## 이미지 슬라이드쇼

```python
import os
```

```python
img_list = os.listdir('./fig/images/')
print(img_list)
```

```python
img_path = []
for i in img_list:
    img_path_all = './fig/images/' + i
    img_path.append(img_path_all)
img_path
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN, # 전체화면
                        cv2.WINDOW_FULLSCREEN)

idx=0
cnt = len(img_path)
while True:
    img_name = img_path[idx]
    img = cv2.imread(img_name, cv2.IMREAD_COLOR)
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(1000) == 27:
        break
    
    idx += 1
    
    if idx >= cnt:
        idx = 0
        
cv2.destroyAllWindows()
```

#### 나의 이미지 슬라이드쇼
